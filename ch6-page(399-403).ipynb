که مطابق شکل ۶-۱۰ (ب) با تناوب $۲\pi$ متناوب است. معادلات $(۶-۱۷)$ و $(۶-۱۸)$ و شکل $(۶-۱۰)$ نشان می‌دهند که فیلترهای پایین‌گذر ایده‌آل کاملاً فرکانس‌گزین‌اند. یعنی تمام فرکانس‌های کوچکتر از فرکانس قطع $\omega_c$ را بدون تضعیف می‌گذرانند و تمام فرکانس‌های باند قطع $(\omega > \omega_c)$ را حذف می‌کنند. همچنین فاز این فیلترها صفر است بنابراین اعوجاج فاز ایجاد نمی‌کنند.  

در بخش $(۶-۲)$ دیدیم که مشخصه فاز غیرخطی می‌تواند مشخصات حوزه زمان یک سیگنال را به شدت تغییر دهد، هرچند اندازه طیف آن تغییر نکند. بنابراین فیلتری با مشخصه اندازه بیان‌شده در معادلات $(۶-۱۷)$ و $(۶-۱۸۶)$، ولی با فاز غیرخطی می‌تواند در بعضی کاربردها اثرات نامطلوبی ایجاد کند. ولی اگر فاز فیلتر ایده‌آل مطابق شکل $(۶-۱۱)$ در باند عبور خطی باشد، نسبت به فیلتر مشابهی با مشخصه فاز صفر تنها یک جابجایی زمانی ایجاد می‌کند.  

در مثال‌های $(۴-۱۸)$ و $(۵-۱۲)$ پاسخ ضربه‌های پایین‌گذر ایده‌آل را به دست آوردیم. پاسخ ضربه فیلتر بیان‌شده با معادله $(۶-۱۷)$ عبارت است از:

$$ h(t) = \frac{\sin(\omega_c t)}{\pi t} \quad \text{(۶-۱۹)} $$

$$ h[n] = \frac{\sin(\omega_c n)}{\pi n} \quad \text{(۶-۲۰)} $$

که در شکل $(۶-۱۲)$ (ب) به ازای $\omega_c = \frac{\pi}{۴}$ نشان داده شده است. اگر پاسخ فرکانسی‌های ایده‌آل معادلات $(۶-۱۷)$ یا $(۶-۱۸)$ با مشخصه فاز خطی در نظر گرفته شوند، پاسخ ضربه تنها یک تأخیر زمانی برابر با منفی شیب تابع فاز پیدا می‌کند، این مطلب در شکل $(۶-۱۳)$ برای حالت پیوسته در زمان نشان داده شده است. توجه کنید که در هر دو حالت پیوسته در زمان و گسسته در زمان، پهنای باند عبور فیلتر $\omega_c$ است ولی پهنای لاله اصلی پاسخ ضربه $\frac{۱}{\omega_c}$ متناسب است. با افزایش پهنای باند، پاسخ ضربه باریک‌تر می‌شود و برعکس، که با رابطه معکوس مشخص می‌شود.

###### پوسته در زمان با مشخصه فاز خطی


حوزه زمان-حوزه فرکانس بیان شده در فصل‌های ۴ و ۵ سازگار است.  
پاسخ پله‌های \( s(t) \) و \( s[n] \) فیلترهای ایده‌آل پیوسته در زمان و گسسته در زمان در شکل ۶-۱۴ نشان داده شده است.  

در هر دو حالت، پاسخ ضربه مشخصاتی دارند که ممکن است مطلوب نباشد. پاسخ پله در هر دو حالت نسبت به مقدار نهایی فراجهش دارد و رفتار نوسانی از خود نشان می‌دهد. این رفتار را **نوسان میرا** می‌نامند.  

می‌دانیم که پاسخ پله انتگرال یا جمع انباره‌ای پاسخ ضربه است، یعنی:  

$$
s(t) = \int_{-\infty}^{t} h(\tau) d\tau
$$

$$
s[n] = \sum_{m=-\infty}^{n} h[m]
$$

چون لُب اصلی پاسخ ضربه فیلترهای ایده‌آل بین \( -\frac{\pi}{\omega_c} \) و \( \frac{\pi}{\omega_c} \) قرار دارد، عمده تغییرات پاسخ پله در این فاصله رخ می‌دهد. پس زمان صعود پاسخ پله، که معیاری از زمان پاسخ فیلتر است، با عکس پهنای باند فیلتر متناسب است.

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترهای فیلتر
wc = 5  # فرکانس قطع
t = np.linspace(-2, 2, 1000)  # محور زمان

# پاسخ ضربه‌ای
h = (wc / np.pi) * np.sinc(wc * t / np.pi)

# رسم نمودار
plt.figure(figsize=(6, 3))
plt.plot(t, h, 'k', linewidth=2)
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.xlabel('t')
plt.ylabel('h(t)')
plt.grid()
plt.show()
```

### ****۶-۴ جنبه‌های حوزه زمان و حوزه فرکانس فیلترهای غیر ایده‌آل****


در عمل مشخصات فیلترهای ایده‌آل همیشه مطلوب نیستند.  
مثلاً در بسیاری از کاربردها، سیگنال‌هایی که باید جدا شوند، در باندهای فرکانسی کاملاً مجزا قرار ندارند.  
یک وضعیت نوعی می‌تواند به صورت شکل ۶-۱۵ .


<img src="https://raw.githubusercontent.com/sepidehrezvani/10-/b75478abb563bd23bf93848a4839a2eb630a24dc/%D8%B4%DA%A9%D9%84%206-15.jpg" width="500"/>


```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف محور فرکانس
omega = np.linspace(-5, 5, 400)

# تابع طیف مثلثی با دامنه کم
def triangular_spectrum(omega):
    return np.maximum(0.3 - np.abs(omega)/6, 0)  # کاهش دامنه

# تابع طیف نوسانی با دامنه کمتر
def wavy_spectrum(omega):
    return np.exp(-((omega + 2)**2)) * (0.15 + 0.1 * np.sin(3 * omega))  # کاهش دامنه

# محاسبه مقدار طیف‌ها
X1 = triangular_spectrum(omega)
X2 = wavy_spectrum(omega)

# رسم نمودار
plt.figure(figsize=(7, 4))
plt.plot(omega, X1, label=r"$X_1(j\omega)$", color="black")
plt.plot(omega, X2, label=r"$X_2(j\omega)$", color="black")
plt.plot(omega, np.maximum(X1, X2), label=r"$X(j\omega)$", color="black", linewidth=2)

# تنظیمات نمودار
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel(r"$\omega$")
plt.ylabel(r"$X(j\omega)$")
plt.legend()
plt.grid(False)
plt.ylim(-0.1, 0.5)  # محدود کردن دامنه محور y
plt.show()
```

نکته دیگر از بررسی پاسخ پله فیلترهای پایین گذر ایده‌آل شکل ۶-۱۴، روشن می‌شود. در هر دو حالت پیوسته در زمان و گسسته در زمان پاسخ پله به مقدار ثابتی برابر اندازه پله میل می‌کند. ولی در حوالی ناپیوستگی فراجهش و نوسان میرا رخ می‌دهد. در بعضی مواقع چنین رفتاری نامطلوب به حساب می‌آید.  

به علاوه حتی در مواردی که مشخصه فرکانس گزینی ایده‌آل امری مطلوب است، دست یافتن به آن نشدنی است. معادلات (۶-۱۸) و (۶-۱۹) و شکل ۶-۱۲ نشان می‌دهند که فیلتر پایین گذر ایده‌آل غیرعلی است. وقتی عمل فیلتر کردن در زمان واقعی صورت می‌گیرد علی بودن قیدی لازم است و مشخصه ایده‌آل باید به صورتی علی تقریب زده شود.  

ملاحظه دیگری که انعطاف در مشخصات فیلتر را می‌طلبد سهولت ساخت است. به طور کلی هر چه بکوشیم فیلتر فرکانس گزین ایده‌آل را دقیق‌تر تقریب بزنیم ساختن آن پیچیده‌تر و پرهزینه‌تر می‌شود چه در حالت پیوسته در زمان از لحاظ عناصری چون مقاومت خازن، القاگر و آپ‌امپ، و چه در حالت گسسته در زمان از لحاظ حافظه ضرب‌کننده و جمع‌کننده. در بسیاری از موارد مشخصه دقیق فیلتر امری اساسی نیست و فیلتری ساده کفایت می‌کند.  

به خاطر تمام دلایل بالا فیلترهای غیر ایده‌آل اهمیت عملی فراوانی دارند و مشخصات این فیلترها در حوزه فرکانس و حوزه زمان با چند پارامتر بیان می‌شوند. اول این که چون مشخصه اندازه فیلتر فرکانس گزین ایده‌آل ممکن است قابل دسترس یا مطلوب نباشد رفتار فیلتر در باندگذار با انعطاف بیشتری تعریف شده، گذر تدریجی از باند عبور به باند قطع مجاز شمرده می‌شود. در مورد فیلتر پایین گذر انحراف از بهره یک در باند عبور و از بهره صفر در باند قطع جایز است و مشخصات لبه‌هایی برای باند عبور و باند قطع و فاصله‌ای بین این دو باند را بیان می‌کنند. وضعیت به صورت نشان داده شده در شکل ۶-۱۶ است و خواسته می‌شود که اندازه پاسخ.

###### شکل ۶-۱۶ تولرانس‌های اندازه فیلتر پایین گذر. تموج مجاز باند $ \delta_1 $ و تموج مجاز باند قطع $ \delta_2 $ است.  
###### منحنی خط‌چین یک پاسخ فرکانسی ممکن است که در محدوده تولرانس‌های مجاز قرار داد.

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترهای پاسخ سیستم مرتبه دوم
zeta = 0.2    # ضریب میرایی (برای نوسانات بیشتر مقدارش کم باشه)
wn = 5        # فرکانس طبیعی
wd = wn * np.sqrt(1 - zeta**2)  # فرکانس نوسانی

# محور زمان
t = np.linspace(0, 4, 500)
s = 1 - np.exp(-zeta * wn * t) * (np.cos(wd * t) + (zeta / np.sqrt(1 - zeta**2)) * np.sin(wd * t))

# رسم نمودار
fig, ax = plt.subplots(figsize=(6, 3.5))  # کوچیک‌تر کردن شکل برای کادر بهتر
ax.plot(t, s, 'k', linewidth=1.8)

# رسم خطوط کمکی
ax.axhline(1, color='gray', linestyle='--', alpha=0.7)
ax.axhline(1 + 0.1, color='gray', linestyle='--', alpha=0.7)  # +Δ
ax.axhline(1 - 0.05, color='gray', linestyle='--', alpha=0.7) # -δ
ax.axvline(0.8, color='black', linestyle='--', alpha=0.7)  # زمان صعود
ax.axvline(3.2, color='black', linestyle='--', alpha=0.7)  # زمان نشست

# نمایش بازه نوسانات (با کمی تغییر مکان)
ax.plot([1.2, 3.2], [1, 1], 'k', linewidth=1.5)
ax.text(2.2, 1.07, r'$ \frac{2\pi}{\omega_d} $', fontsize=11, bbox=dict(facecolor='white', edgecolor='none', alpha=0.7))

# برچسب‌گذاری
ax.text(0.7, -0.05, r'$t_r$', fontsize=10)
ax.text(3.1, -0.05, r'$t_s$', fontsize=10)
ax.text(0.1, 1.12, r'$\Delta$', fontsize=10)
ax.text(0.1, 0.92, r'$\delta$', fontsize=10)
ax.set_xlabel('t', fontsize=11)
ax.set_ylabel('s(t)', fontsize=11)

# تنظیمات نمایش برای قرارگیری در کادر
ax.set_ylim(-0.1, 1.2)
ax.set_xlim(0, 4)
ax.set_xticks(np.arange(0, 4.5, 1))
ax.set_yticks(np.arange(0, 1.3, 0.2))
ax.grid(True, linestyle='--', alpha=0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout(pad=1.5)  # تنظیم خودکار برای جا شدن در کادر
plt.show()
```


###### شکل ۶-۱۷ پاسخ پله یک فیلتر پایین گذر پیوسته در زمان که در آن زمان صعود $ t_r $، فراجهش $ \Delta $، فرکانس نوسان‌های میرا $ \omega_r $ و زمان نشست $ t_s $، یعنی زمان لازم برای قرار گرفتن پاسخ پله در فاصله$ ± \delta $ مقدار نهایی، نشان داده شده است.


فرکانسی در ناحیه غیر خاکستری قرار گیرد. در این شکل اجازه داده شده که بهره در باند عبور به اندازه $ \delta_1 $ از یک  
و در باند قطع به اندازه $ \delta_2 $ از صفر فاصله بگیرد.  

تفاوت اندازه پاسخ فرکانسی در باند عبور از یک را **تموج باند عبور** و تفاوت اندازه پاسخ فرکانسی در باند  
قطع از صفر را **تموج باند قطع** می‌نامند. $ \omega_p $ لبه باند عبور و $ \omega_s $ لبه باند قطع نامیده می‌شوند.  
گستره فرکانسی بین $ \omega_p $ و $ \omega_s $ برای گذار از باند عبور به باند قطع است و **باند گذار** خوانده می‌شود.  
برای فیلترهای پایین‌گذر گسسته در زمان و دیگر فیلترهای فرکانس‌گزین پیوسته در زمان و گسسته در زمان نیز تعاریف مشابهی وجود دارد.  

علاوه بر مشخصات اندازه در **حوزه فرکانس**، **مشخصه فاز** نیز در مواردی مهم است.  
خصوصاً **خطی یا تقریباً خطی بودن فاز** در باند عبور معمولاً امری مطلوب است.  

برای کنترل **رفتار حوزه زمان** معمولاً قیودی روی **پاسخ پله فیلتر** گذاشته می‌شود.  
چنانچه **شکل ۶-۱۷** نشان می‌دهد، یکی از کمیت‌های موردنظر **زمان صعود** $ t_r $ پاسخ پله است.  
زمان صعود فاصله‌ای را بیان می‌کند که در آن **پاسخ پله به مقدار نهایی‌اش می‌رسد**.  
بود یا نبود **رفتار نوسانی** نیز غالباً مهم است.  

در صورت **وجود نوسان میرا**، سه کمیت دیگر نیز برای بیان رفتار آن بیان می‌شود:  
- **فراجهش** $ \Delta $ نسبت به مقدار نهایی پاسخ پله  
- **فرکانس نوسان‌ها** $ \omega_r $  
- **زمان نشست** $ t_s $، زمانی است که طول می‌کشد تا پاسخ پله در محدوده خاصی از مقدار نهایی قرار گیرد.  

در **فیلترهای پایین‌گذر غیر ایده‌آل** بین **عرض ناحیه گذار** (که مشخصه‌ای از حوزه فرکانس است)  
و **زمان نشست پاسخ پله** (که مشخصه‌ای از حوزه زمان است) **بده‌بستانی** وجود دارد.  
**مثال زیر این بده‌بستان را نشان می‌دهد.**
